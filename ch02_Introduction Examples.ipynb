# 2. 사례 소개

### 이번장 주의 사항

- 이번장은 이 책을 활용해서 어떤 데이터를 분석할 수 있는지 보여주는 사례 중심으로 설명한다. 그렇기 때문에 나같은 초보자들은 맨붕을 당할 수 밖에 없다. 나도 초반에 이 책을 따라한답시고 처음부터 했는데 내가 코드를 치고 있는데도 불구하고 이게 어떤 역할을 하는 코드인지 알 수 없었다. 그러므로 이번 장은 가벼운 마음으로 한 번 훑어본다는 생각을 갖자.
- 모든 장을 공부하고 다시 2장으로 돌아와서 코드를 쳐본다면 달라진 자신을 느낄 수 있을 것이다.
- 이 책은 좀 더 세련된 분석 방법에 대한 설명이 아닌, **도구에 초점**을 두고 있다. 
- 이 책에서 **소개하는 기술을 익히고 나면 원하는 것이 무엇인지 명확하다면 자신만의 분석**을 조속히 구현

## 파이썬 사용하는 목적

- 외부 자료 활용: 여러 가지 파일 형식과 데이터베이스를 읽고 쓰는 작업
- 데이터 준비: 데이터 정비, 수집, 집계, 정규화, 개조, 분리, 분석을 위한 데이터 변형
- 데이터 변형: 모아놓은 데이ㅓ에 수학적, 통계학적 연산을 적용해 새로운 데이터를 도출. 예) 그룹 변수로 분류된 큰 표 만들기
- 모델링과 계산: 통계 모델, 기계 학습 알고리즘 또는 다른 계산 도구와 데이터를 연결하기
- 데이터 표현: 정적 혹은 인터랙티브한 도식화 또는 원문 요약
